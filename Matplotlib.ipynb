```python
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.title("Sine and Cosine waveform using Matplotlib")
plt.plot(x, y_sin, 'm')
plt.plot(x, y_cos, 'c')
```

```python
# Bar Plot

x1 = [2, 8, 15]
y1 = [5, 10, 3]

x2 = [1, 7, 5]
y2 = [12, 15, 9]

plt.bar(x1, y1)
plt.bar(x2, y2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sample Bar Graph')
plt.legend(labels = ['Feature 1', 'Feature 2'])
```

```python
# Box Plot

data = [np.random.randn(100)]   # Returns a sample from the “standard normal” distribution

plt.boxplot(data, vert=True, patch_artist=True)
```

```python
plt.boxplot(data, vert=False, patch_artist=True)
```

```python
plt.boxplot(data, vert=True, patch_artist=False)
```

```python
# Pie Charts

labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0.01, 0.01, 0.01)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)
plt.axis('equal')
```
