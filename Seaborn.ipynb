```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
# Finding correlation between numerical features

df.corr()
```

```python
# Visualizing correlation matrix using a heatmap

sns.heatmap(df.corr())
```

```python
sns.heatmap(df.corr(), annot=True) # turning on the annotations
```

```python
# JointPlot - used for studying relationship between two numerical variables

sns.jointplot(x='tip', y='total_bill', data=df) # by default draws a scatterplot with histograms
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex') # replaces the scatterplot with hexagonal bins
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg') # adds regression and kernel density fits
```

```python
# Pair plots - plots pairwise relationships in a dataset

sns.pairplot(df)  # Draw scatterplots for joint relationships and histograms for univariate distributions
```

```python
sns.pairplot(df, hue='sex')  # Show different levels of a categorical variable by the color of plot elements
```

```python
sns.pairplot(df, hue='sex', palette='husl')  # Using a different color palette
```

```python
# Dist plots - plot univariate distribution of observations (similar to hist() in Matplotlib)

sns.distplot(df['tip'])  # by default, shows kernel density estimate (kde) and histogram 
                         # with bin size determined aytomatically with a reference rule
```

```python
sns.distplot(df['tip'], kde=False)  # only shows histogram bins
```

```python
sns.distplot(df['tip'], rug=True, hist=False) # plot distribution with kde and rug plot
```

```python
sns.distplot(df['tip'], vertical=True)  # plot ditribution on vertical axis
```

```python
# Count plots - shows the counts of observations in each categorical bin using bars

sns.countplot(x='sex', data=df)  # value counts of a single categorical variable
```

```python
sns.countplot(y='sex', data=df)  # plot the bars horizontally
```

```python
sns.countplot(x='sex', hue='smoker', data=df)  # show value counts for two categorical variables
```

```python
sns.countplot(x='sex', hue='smoker', data=df, palette='Pastel2')  # use a different color palette
```

```python
# Bar plots - represents an estimate of central tendency(by default, mean) for a numerical variable based
#             on a categorcal variable

sns.barplot(x='day', y='total_bill', data=df, palette='rainbow')  # Draw a set of vertical bar plots grouped 
                                                                  # by a categorical variable
```

```python
sns.barplot(y='day', x='total_bill', data=df, palette='rainbow') # Draw a set of horizontal bars
```

```python
sns.barplot(x='day', y='total_bill', data=df, hue='sex', palette='Pastel1')  # Draw a set of vertical bars with 
                                                                             # nested grouping by two variables
```

```python
from numpy import median
sns.barplot(x='day', y='total_bill', data=df, palette='rainbow', estimator=median) # use median as the estimate of
                                                                                   # central tendency
```

```python
sns.barplot(x="day", y="tip", data=df, ci="sd", palette='Pastel2') # Show standard deviation of observations
                                                                   # instead of a confidence interval
```

```python
sns.barplot(x="day", y="tip", data=df, palette='Pastel1', capsize=.2)  # Add caps to the error bars
```

```python
# Box plots - displays distribution of quantitative data. Box shows the quartiles of the dataset, while the whisker
            # extend to show the rest of the distribution. Also gives information about outliers in the dataset

sns.boxplot(x=df['total_bill']) # Draws a single horizontal boxplot
```

```python
sns.boxplot(x='day', y='total_bill', data=df)  # Draws a vertical boxplot grouped by a categorical variable
```

```python
sns.boxplot(x='day', y='total_bill', data=df, hue='smoker', palette='PuRd') # Draws a boxplot with nested 
                                                                               # grouping by two categorical 
                                                                               # variables
```

```python
sns.boxplot(data=df, orient='h', palette='Pastel1') # Draws a boxplot for each numeric variable in a DataFrame
```

```python
# Violin plots - combination of boxplot and kernel density estimate(kde)

sns.violinplot(x='total_bill', data=df)  # Draws a single horizontal violinplot
```

```python
sns.violinplot(y="day", x="total_bill", data=df, palette='rainbow') # Draws a horizontal violinplot grouped by a 
                                                                    # categorical variable
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", data=df, palette="PuRd") # Draw a violinplot with nested
                                                                               # grouping by two categorical 
                                                                               # variables
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",                   # Draws split violins to compare 
                                                                        # across the hue variable and 
                                                                        # quartiles as horizontal lines instead of 
                                                                        # a mini-box
               data=df, palette="PuRd", inner='quartile', split=True)
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", 
               data=df, palette="BuGn", inner='stick', split=True) # Show each observation with a stick inside the 
                                                                   # violin
```
